### 1. Statistical Tests: Runtime and Chromatic Number

```python
from scipy.stats import ttest_rel, wilcoxon
import pandas as pd
import numpy as np

# Drop NaNs
runtime_orig = df["Q_GNN50_Time_s"].dropna()
runtime_imp = df["Q_GNN50_IMP_Time_s"].dropna()
k_orig = df["Q_GNN50_k*"].dropna()
k_imp = df["Q_GNN50_IMP_k*"].dropna()

# Runtime comparison
t_stat_rt, p_val_rt = ttest_rel(runtime_orig, runtime_imp)
w_stat_rt, p_val_w_rt = wilcoxon(runtime_orig, runtime_imp)

# Chromatic number comparison
t_stat_k, p_val_k = ttest_rel(k_orig, k_imp)
w_stat_k, p_val_w_k = wilcoxon(k_orig, k_imp)

print("=== Runtime Comparison ===")
print(f"Paired t-test: t = {t_stat_rt:.4f}, p = {p_val_rt:.4g}")
print(f"Wilcoxon signed-rank: W = {w_stat_rt}, p = {p_val_w_rt:.4g}\n")

print("=== Chromatic Number Comparison ===")
print(f"Paired t-test: t = {t_stat_k:.4f}, p = {p_val_k:.4g}")
print(f"Wilcoxon signed-rank: W = {w_stat_k}, p = {p_val_w_k:.4g}")
```

## Deterministic vs GNN

```python
df_common = df.dropna(subset=["GCA_k*", "GCMO_k*", "Q_GNN50_k*"])
from scipy.stats import ttest_rel, wilcoxon

gca_k = df_common["GCA_k*"]
gcmo_k = df_common["GCMO_k*"]
gnn_k = df_common["Q_GNN50_k*"]

# Paired t-tests
t_gca, p_gca = ttest_rel(gca_k, gnn_k)
t_gcmo, p_gcmo = ttest_rel(gcmo_k, gnn_k)

# Wilcoxon tests
w_gca, pw_gca = wilcoxon(gca_k, gnn_k)
w_gcmo, pw_gcmo = wilcoxon(gcmo_k, gnn_k)

print("Chromatic Number Comparison (k*)")
print(f"GCA vs GNN: t={t_gca:.4f}, p={p_gca:.4g} | Wilcoxon p={pw_gca:.4g}")
print(f"GCMO vs GNN: t={t_gcmo:.4f}, p={p_gcmo:.4g} | Wilcoxon p={pw_gcmo:.4g}")
import matplotlib.pyplot as plt

# Calculate average chromatic numbers
avg_k = {
    "GCA": gca_k.median(),
    "GCMO": gcmo_k.median(),
    "Q_GNN50": gnn_k.median()
}

plt.figure(figsize=(8, 5))
plt.bar(avg_k.keys(), avg_k.values(), color=["skyblue", "lightgreen", "salmon"])
plt.title("Average Chromatic Number (k*) Comparison")
plt.ylabel("Average k*")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
